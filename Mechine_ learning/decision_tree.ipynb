```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.25,random_state=42)
```

```python
y_pred=model.predict(x1_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("acc:",accuracy_score(y1_test,y_pred))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
# Step 1: get probability of class 1
y_prob=model.predict_proba(x1_test)[:,1]

# Step 2: compute ROC values
fpr,tpr, thresholds=roc_curve(y1_test,y_prob)

# Step 3: compute AUC
auc=roc_auc_score(y1_test,y_prob)

# Step 4: plot
plt.plot(fpr,tpr,label="AUC:" + str(round(auc, 3)))
plt.plot([0,1],[0,1], linestyle= "--")
plt.xlabel("fpr")
plt.ylabel("tpr")
plt.title("roc curve")
plt.legend()
plt.show()
```
