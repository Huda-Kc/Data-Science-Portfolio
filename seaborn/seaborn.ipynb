```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
tips=sns.load_dataset('tips')
tips.head()

```

```python
sns.kdeplot(x='total_bill',hue= 'time' ,fill=True,alpha=0.6, linewidth=1.5, data=tips)
plt.title('Density curve')
plt.xlabel('Total Bill ($)')
plt.ylabel('Density')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',hue='time',   data=tips,palette='Set2',linewidth=1.5,fliersize=4)
plt.title('Box plot')
plt.xlabel('Day of the week')
plt.ylabel('Total bill')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
subjects=['Math','sci','Eng']
marks=[85,90,98]

plt.bar(subjects,marks)
plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('scatter plot')
plt.show()
```

```python
plt.barh(subjects,marks)
plt.show()
```

```python
plt.scatter(x,y)
plt.title('Scatter plot')
plt.show()
```
