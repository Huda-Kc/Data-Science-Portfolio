```python
# outliers

Q1=df['avg_purchase_value'].quantile(0.25)
Q2=df['avg_purchase_value'].quantile(0.50)
Q3=df['avg_purchase_value'].quantile(0.75)

print(Q1,Q2,Q3)

IQR=Q3-Q1
print(IQR)

upper=Q3+1.5*IQR
print(upper)
lower=Q1-1.5*IQR
print(lower)

outlier=[]
for x in df['avg_purchase_value']:
    if((x>upper or (x<lower))):
        outlier.append(x)
outlier        
```

```python
df=df[(df['avg_purchase_value']>=lower) & (df['avg_purchase_value']<=upper)]
```

```python
Q1=df['churned'].quantile(0.25)
Q2=df['churned'].quantile(0.50)
Q3=df['churned'].quantile(0.75)

print(Q1,Q2,Q3)

IQR=Q3-Q1
print(IQR)

upper=Q3+1.5*IQR
print(upper)
lower=Q1-1.5*IQR
print(lower)

outlier=[]
for x in df['churned']:
    if((x>upper or (x<lower))):
        outlier.append(x)
outlier        
```

```python
df=df[(df['churned']>=lower) & (df['churned']<=upper)]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
# Day 5: Bivariate analysis (scatterplots, correlations, groupby analysis)

# Select numerical columns and calculate correlation
corr_matrix = df[['age','annual_income','spending_score','num_purchases','membership_years']].corr()

sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("correlation heatmap")
plt.show()
```

```python
sns.scatterplot(data=df,x='annual_income',y='spending_score',hue='churned',alpha=0.5)
plt.title("income vs spending score (colored by churn)")
plt.show()
```

```python

# Which country has the highest average spending?
country_spending=df.groupby('country')['avg_purchase_value'].mean().sort_values(ascending=False)
# Which gender is more likely to churn?
gender_churn =df.groupby('gender')['churned'].mean()
print(country_spending)
print(gender_churn)
```

```python
#  Advanced exploration (feature engineering, trends, refined dataset)
df['conversion_rate']=df['num_purchases']/df['website_visits_per_month']
df['conversion_rate']
```

```python
df['age_group'] = pd.cut(df['age'],bins=[0,30,50,100], labels=['Young','Middle_aged','Senior'])
df['age_group']
```

```python
df['last_purchase_date'] = pd.to_datetime(df['last_purchase_date'])
df['days_last_purchase'] =(pd.to_datetime('today') - df['last_purchase_date']).dt.days
df['days_last_purchase']
```

```python
plt.figure(figsize=(12,7))

sns.scatterplot(data=df,x='days_last_purchase',y='conversion_rate',hue='country',s=100)
plt.axhline(y=df['conversion_rate'].mean(),color='red',linestyle='--',label='average conversion')
plt.axvline(x=df['days_last_purchase'].median(),color='blue',linestyle='--',label='median days')

plt.title('Recency vs Conversion Rate Analysis', fontsize=15)
plt.xlabel('Days Since Last Purchase', fontsize=12)
plt.ylabel('Conversion Rate', fontsize=12)
plt.legend()
plt.show()
```
